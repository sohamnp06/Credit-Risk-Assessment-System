```python
df["Monthly_Income"] = df["Income"] / 12
```

```python
r = (df["InterestRate"] / 100) / 12
n = df["LoanTerm"]

df["Estimated_EMI"] = (
    df["LoanAmount"] * r * (1 + r) ** n
) / ((1 + r) ** n - 1)
```

```python
df["EMI_to_Income"] = df["Estimated_EMI"] / df["Monthly_Income"]
```

```python
df["Disposable_Income"] = df["Income"] * (1 - df["DTIRatio"])
```

```python
df["Income_After_EMI"] = (
    df["Monthly_Income"] - df["Estimated_EMI"]
)
```

```python
bins = [df.CreditScore.min()-1,
        580,
        670,
        740,
        800,
        df.CreditScore.max()+1]
labels = [
    "Poor",
    "Fair",
    "Good",
    "Very Good",
    "Excellent"
]

df["CreditScore_Bucket"] = pd.cut(
    df["CreditScore"],
    bins=bins,
    labels=labels
)
```

```python
df["Income_Group"] = pd.qcut(
    df["Income"],
    q=4,
    labels=[
        "Low",
        "Medium",
        "High",
        "Very High"
    ]
)
```

```python
df["Employment_Stability"] = (
    df["MonthsEmployed"] / df["Age"]
)
```

```python
df["Loan_Burden"] = (
    df["Loan_to_Income"] *
    df["DTIRatio"]
)
```

```python
df.to_csv('../data/da-loanDefaulter.csv', index=False)
```
